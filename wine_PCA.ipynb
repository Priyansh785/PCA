```python
# Converting data to numpy array
data_ary=df1.values
data_ary
```

```python
from sklearn.preprocessing import scale
```

```python
# Normalizing the  numerical data
data_norm=scale(data_ary)
data_norm
```

```python
from sklearn.decomposition import PCA
```

```python
# Applying PCA Fit Transform to dataset
pca = PCA()
pca_values = pca.fit_transform(data_norm)
pca_values
```

```python
# PCA Components matrix or convariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var,color="red");
```

```python
# Final Dataframe
df2=pd.concat([data['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
df2
```

```python
# Visualization of PCAs

fig=plt.figure(figsize=(12,8))
sns.scatterplot(data=df2);
```

## Checking with other Clustering Algorithms

### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(data_norm,'complete'))
```

```python
# Create Clusters
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(data_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
df3=data.copy()
df3['clustersid']=hclusters.labels_
df3
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# within-cluster sum-of-squares criterion 

wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

### Build Cluster algorithm using K-3

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(data_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
df4=data.copy()
df4['clusters3id']=clusters3.labels_
df4
```
